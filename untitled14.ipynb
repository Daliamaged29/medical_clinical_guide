Loading data

```python
from PyPDF2 import PdfReader
reader=PdfReader("guideline-170-en_نسخة (1).pdf")
text=""
for page in reader.pages:
  text+=page.extract_text()
print(text[:300])
```

Cleaning data

```python
import re
def clean_medical_text(text):
  text=re.sub(r"Page\s*\d+\s*(/|of)\s*\d+Exported\s*on:\s*\d{2}/\d{2}/\d{4}","",text,flags=re.IGNORECASE)
  text=re.sub(r"^\d+$","",text,flags=re.MULTILINE)
  text=re.sub(r"all\s*rights\s*reserved.*?\.","",text)
  text=re.sub(r"\n\s*\n","\n",text)
  return text.strip()
text_cleaned=clean_medical_text(text)
print(text_cleaned[:300])
```

Split

```python
text_cleaned="".join(text_cleaned.split())
print(text_cleaned[:300])
```

Chunks

```python
chunk_size=1000000
chunk_overlap=500
chunks=[text_cleaned[i:i+chunk_size] for i in range(0,len(text_cleaned),chunk_size-chunk_overlap)]
print(len(chunks))
print(chunks[0])
```

Embedding

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer("all-MiniLM-L6-v2")
embeddings=model.encode(chunks)
```

```python
print(embeddings.shape)
print(embeddings[0])
```

Question

```python
question="How to prevent malaria?"
```

RAG Pipeline

```python
query_embedding=model.encode([question])
```

Retrieval

```python
from sklearn.metrics.pairwise import cosine_similarity
best_chunk=chunks[cosine_similarity(query_embedding,embeddings).argmax()]
```

```python
prompt=f"""
Answer the question using only the following context.

Context:
{best_chunk}

Question:
{question}
"""
```

LLM & Generate

```python
from google import genai
from google.colab import userdata
api_key=userdata.get("GEMINI_API_KEY")
client=genai.Client(api_key=api_key)

response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```
