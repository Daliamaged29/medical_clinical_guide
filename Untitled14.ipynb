```python
import re
def clean_medical_text(text):
  text=re.sub(r"Page\s+\d+\s*(/|of)\s*\d+","",text,flags=re.IGNORECASE)
  text=re.sub(r"\d+$","",text,flags=re.MULTILINE
  text=re.sub(r"all rights reserved.*?\.","",text)
  text=re.sub(r"\n\s*\n","\n",text)
  text=text.script()
  return text
  print(text[:300])
```
